# **Q2 A)**

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random

def poiss(lam,x,t):
    a=lam*t
    b=a**x
    c=math.exp(-a)
    d=math.factorial(x)
    return (b*c)/d

lam=5
time=[3,5]
for t in time:
  x=[i for i in range(1,100)]
  y=[poiss(lam,i,t) for i in x]
  plt.figure(figsize=(10, 5))
  plt.plot(x, y, color='blue', label=f'Density (t={t})')
  plt.xlabel(f'Number of Arrivals in time t={t} (5 hours)')
  plt.ylabel('Density')
  plt.title(f'Density Plot (Poisson, =5, t={t})')
  plt.grid(axis='y', linestyle='--', alpha=0.7)
  plt.legend()
  plt.show()

  cumulative_frequencies = np.cumsum(y)

  plt.figure(figsize=(10, 5))
  plt.plot(x, cumulative_frequencies, marker='o', linestyle='-', color='green', label=f'CDF (t={t})')
  plt.xlabel(f'Number of Arrivals in time t={t} (5 hours)')
  plt.ylabel('Cumulative Frequency (P X <= x)')
  plt.title(f'Cumulative Distribution Function (CDF) (Poisson, =5, t={t})')
  plt.grid(axis='y', linestyle='--', alpha=0.7)
  plt.legend()
  plt.show()
```

# **Q2 C)**

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random

def expo1(lam,t):
    a=lam*t
    b=math.exp(-a)
    c=1-b
    return c

def expo2(lam,t):
    a=math.exp(-lam*t)
    b=lam*a
    return b

l=[5,15]

for lam in l:

  t=[i for i in np.arange(0,3,0.1)]
  y=[expo2(lam,i) for i in t]

  plt.figure(figsize=(8, 6))
  plt.plot(t, y, label='Density: $\lambda e^{-\lambda t}$', color='green')
  plt.xlabel('Time (t)')
  plt.ylabel('Density')
  plt.title('PDF for first-interarrival of the process ')
  plt.legend()
  plt.show()

  #increment t by 0.1
  t=[i for i in np.arange(0,3,0.1)]
  y=[expo1(lam,i) for i in t]

  plt.figure(figsize=(8, 6))
  plt.plot(t, y, label='Density: $1 - e^{-\lambda t}$', color='blue')
  plt.xlabel('Time (t)')
  plt.ylabel('Density')
  plt.title('CDF for first-interarrival time')
  plt.legend()
  plt.show()






```
